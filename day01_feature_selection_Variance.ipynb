```python
import pandas as pd
from sklearn.feature_selection import VarianceThreshold
```

```python
# 读取数据
data = pd.read_csv("../data/factor_returns.csv")
data
```

```python
# 选择数据
data = data.iloc[:, 1:-2]
data
```

```python
# 实例化一个转换器类
transfer = VarianceThreshold(threshold=10)
# 调用fit_transfer
data_new = transfer.fit_transform(data)
data_new
```
