```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 设置热力图默认样式
sns.set()
# 设置热力图样式
sns.set_style("whitegrid", {"font.sans-serif":["simhei", "Arial"]})
```

```python
# 生成随机的二维数组
data = pd.DataFrame(
    np.random.rand(4, 7),
    index = ["天安门","故宫","颐和园","八达岭长城"],
    columns = ["周一","周二","周三","周四","周五","周六","周日"]
)
data
```

```python
data.describe()
```

```python
# 画热力图
plt.figure(figsize=[10, 4])
# sns.heatmap(data, annot="True", fmt=".2f", cmap="coolwarm")
# fmt:数据格式，annot:是否显示数字，cmap:
sns.heatmap(data, fmt=".2f", annot=True, cmap="coolwarm")
```

```python
# 读取坦坦尼克号的数据
data = pd.read_csv("../data/titanic.csv")
data
```

```python
# 数据选择
data = data.iloc[:, 1:]
data
```

```python
# 查看数据类型
data.info()
```

```python
# pands把字符串类型的列，变为分类数字编码
for field in ["sex", "boat", "embarked"]:
    data[field] = data[field].astype("category").cat.codes
```

```python
# 查看数据
data
```

```python
# 查看数据类型
data.info()
```

```python
# 计算相关系数
data.corr()
```

```python
# 画热力图
plt.figure(figsize=[10, 4])
# fmt:数据格式，annot:是否显示数字，cmap:热力图样式
sns.heatmap(data.corr(), annot=True, fmt=".2f", cmap="RdYlGn")
```
