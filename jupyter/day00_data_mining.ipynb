```python
import pandas as pd
left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                        'key2': ['K0', 'K1', 'K0', 'K1'],
                        'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3']})
left
```

```python
right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                        'key2': ['K0', 'K0', 'K0', 'K0'],
                        'C': ['C0', 'C1', 'C2', 'C3'],
                        'D': ['D0', 'D1', 'D2', 'D3']})
right
```

```python
pd.merge(left, right, how="inner", on=["key1", "key2"])
```

```python
stock = pd.read_csv("../data/data_mining/stock_day.csv")
stock
```

```python
temp_stock[temp_stock["p_change"] > 2].head() 
```

```python
temp_stock[temp_stock > 25] = 2
temp_stock
```

```python
temp_stock.sort_values(by="high", ascending=True)
```

```python
temp_stock2 = np.where(temp_stock>20, 1, 0)
temp_stock2
```

```python
stock["pona"] = np.where(stock["p_change"] > 0, 1, 0)
stock.head()
```

```python
date = pd.to_datetime(stock.index)
date
```

```python
stock["week"] = date.weekday
stock
```

```python
data =  pd.crosstab(stock["week"], stock["pona"])
data
```

```python
temp1 = data.sum(axis = 1)
temp1
```

```python
temp2 = pd.DataFrame({"a":temp1.values, "b":temp1.values})
temp2
```

```python
data.values / temp2.values
```

```python
data.div(data.sum(axis=1), axis =0).plot(kind="bar", stacked=True)
```

```python
movie = pd.read_csv("../data/data_mining/IMDB-Movie-Data.csv")
movie.head()
```

```python
movie["Rating"].mean()
```

```python
np.unique(movie["Director"]).size
```

```python
movie_genre = [i.split(",") for i in movie["Genre"]]
movie_genre
```

```python
movie_class = np.unique([j for i in movie_genre for j in i])
movie_class
```

```python
count = pd.DataFrame(np.zeros(shape=[1000, 20]), dtype=np.int32, columns=movie_class)
count
```

```python
count.loc[0,movie_genre[0]]
```

```python
for i in range(1000):
    count.loc[i,movie_genre[i]] = 1
count
```

```python
count.sum(axis = 0).sort_values(ascending=False).plot(kind="bar", figsize=(20, 9), fontsize=40)
```
